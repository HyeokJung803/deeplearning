### 역전파는 무엇인가?
- 역전파란 신경망이 입력을 받아 예측한 값을 정답과 비교하여 오차를 계산한 것을 거꾸로 추적하여 각 뉴런과 가중치를 분석, 수정한다

### 역전파에서 연쇄 법칙이 필요한 이유는?
- 다층 퍼셉트론에서 오차가 여러 은닉층를 거쳐 전달되어 합성함수이므로, 가중치를 수정할 방향과 크기를 구하기 위해 손실함수를 미분하도록 연쇄 법칙을 사용한다.

### 역전파는 매 epoch마다 몇 번 실행될까?


- Epoch는 전체 학습 데이터를 한 번 모두 사용하여 학습하는 과정을 의미 역전파는 일반적으로 각 배치(batch)마다 1회 수행되므로, 1 Epoch 동안의 역전파 횟수는 전체 데이터 수를 배치 크기로 나눈 값과 같다.

### ReLU를 은닉층에 쓰는 이유는?
- 과거에는 시그모이드 함수를 많이 사용했지만, 은닉층이 많아지면 미분값이 계속 작아서 기울기가 거의 0이 되버리는 문제가 발생
- 하지만 ReLU는 계산이 간단하고 입력이 0보다 크면 그대로 출력, 0 이하면 0을 출력하는 기울기가 잘 유지되어 활용도가 높다

### Adam이 SGD와 다른 점은?
- Adam과 SGD는 둘다 오차를 줄이기 위해 가중치를 수정하는 역할입니다.
- SGD는 현재 기울기만 사용하여, 즉흥적으로 수정하며, Adam은 과거정보를 반영하여 평균값을 내어 안정성이 높다. 

### learning rate가 너무 크면 어떤 문제가 생길까?
- 학습률이 너무 크면 최소점을 향해 내려갈 때 완만하게 내려가지 못하고 최소점을 지나쳐버리는 경우가 발생
- 오히려 최솟값에서 멀어지는 발산상태가 된다
- 오버슈팅이 반복되면 가중치 값이 기하급수적으로 커지게 되는 상황 발생

### Flatten() 레이어의 파라미터 수는?
- 0개
- flatten는 입력받은 다차원 배열을 1차원 형태로 바꾸는 역할만 하기 때문이다.

```python
import tensorflow as tf
from tensorflow.keras import layers, models

# 1. 데이터셋 로드 (동일한 Fashion-MNIST 사용)
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()

# 2. 전처리 (0~1 사이 값으로 정규화)
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

# 3. 모델 구축 (Flatten 레이어 포함)
model = models.Sequential([
    # (28, 28)의 2차원 이미지를 784개의 1차원 배열로 펼쳐줌 (파라미터 수: 0개)
    layers.Flatten(input_shape=(28, 28)),
    
    # 은닉층 (Hidden Layer)
    layers.Dense(128, activation='relu'),
    
    # 출력층 (10개 의류 카테고리 분류)
    layers.Dense(10, activation='softmax')
])

# 4. 모델 컴파일 (당시 사용했던 RMSprop 옵티마이저 설정)
model.compile(
    optimizer='rmsprop',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# 5. 모델 학습 (훈련 데이터의 20%를 검증 데이터로 분리하여 val_accuracy 확인)
print("--- 모델 학습 시작 ---")
history = model.fit(
    x_train, y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.2  # 실시간으로 val_accuracy를 계산하기 위해 설정
)

# 6. 최종 테스트 데이터 평가 (실전 결과 확인)
print("\n--- 최종 테스트 데이터 평가 ---")
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"\n최종 검증 정확도 (val_accuracy): {test_acc * 100:.2f}%")
```
